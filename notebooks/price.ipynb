```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN, KMeans
from sklearn.neighbors import NearestNeighbors
```

```python
df = pd.read_csv('../barca.csv', header=0)
df.head()
```

```python
os = df['overall_satisfaction'].value_counts()
fig = os.plot.bar()
fig.get_figure().savefig('distribution')
```

```python
coords = df[['latitude', 'longitude']]
nbrs = KMeans(n_clusters=10).fit_predict(coords)
```

```python
plt.scatter('latitude', 'longitude', data=coords, c=nbrs, alpha=0.5, s=3)
```

```python
ax = df.plot(x='price', y='overall_satisfaction', kind='scatter', alpha=0.4)
ax
```

```python
ax = plt.scatter(x=nbrs, y=df['overall_satisfaction'])
```

```python
ds = pd.cut(df['price'], 15, labels=False)
ax = plt.scatter(x=ds, y=df['overall_satisfaction'], alpha=0.4)
ax.set_xlabel('price')
```
